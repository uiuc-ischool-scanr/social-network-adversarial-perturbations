```python
import json
import networkx as nx
import matplotlib.pyplot as plt
from networkx.readwrite import json_graph
from pprint import pprint
import random

folder_path = '../output-experiments/GRAPH_cellular_CENTRALITY_betweenness_DIRECTION_decrease_TIER_10_REACHTYPE_1/'
file_name = 'adv_learning_results_decrease_betweenness_centrality_graphtype_cellular_budget_1_bootstrapsample_0_tier_0_subsample_0.json'

combined_path = folder_path + file_name
with open(combined_path) as the_file:
    adversarial_sequences = json.load(the_file)

if (adversarial_sequences['pre_experiment_metrics']['starting_graph'] == adversarial_sequences['post_experiment_metrics']['ending_graph']):
    print('pre and post graph are the same')
else:
    print('pre and post graphs are different')

print('PRE GRAPH')
pre_graph = json_graph.node_link_graph(adversarial_sequences['pre_experiment_metrics']['starting_graph'])
nx.draw_spring(pre_graph, with_labels=True)
plt.show()

print('POST GRAPH')
post_graph = json_graph.node_link_graph(adversarial_sequences['post_experiment_metrics']['ending_graph'])
nx.draw_spring(post_graph, with_labels=True)
plt.show()

print('Starting_eval')
print(adversarial_sequences['pre_experiment_metrics']['starting_eval'])
print('\n')

print('Starting Percentile')
print(adversarial_sequences['pre_experiment_metrics']['starting_percentile'])
print('\n')

print('Direction')
print(adversarial_sequences['configurations']['direction'])
print('\n')

print('Eval_metric')
print(adversarial_sequences['configurations']['eval_metric'])
print('\n')

print('Self_node')
print(adversarial_sequences['pre_experiment_metrics']['self_node'])
print('\n')

print('Moves Taken')
print(adversarial_sequences['post_experiment_metrics']['optimal_sequence']['details'])
details_moves_key = adversarial_sequences['post_experiment_metrics']['optimal_sequence']['key']
print('\n')

print('Ending_eval')
print(adversarial_sequences['post_experiment_metrics']['ending_eval'])
print('\n')

print('Ending Percentile')
print(adversarial_sequences['post_experiment_metrics']['ending_percentile'])
print('\n')

print('Move Details')
pprint(adversarial_sequences['experiment']["perturbations"][str(details_moves_key)])
```
